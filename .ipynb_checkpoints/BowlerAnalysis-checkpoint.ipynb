<b>Top Wicket Takers</b>
<br>
It is interesting to see how the list of top wicket takers have an equal number of fast bowlers and spinners. Another thing to note here is how CSK dominates this list with 3 bowlers, 2 out of which (Imran Tahir and Harbhajan Singh) are fairly experienced campaigners of this format, this is in contrast to the Top Scorers List where none of their palyers made an appearence. All the other teams also make an appearence here except KKR, the situation for KKR is looking pretty rough as none of their bowlers make an appearence even in Top 20 Wicket Takers. After such a great season both Khaleel Ahmed and Deepak Chahar got the oppurtunity to represent India on international level more frequently, Rabada also had a sensational season with Delhi Capitals and looked cut above others.

```python
def get_wicket_count(player):
    filt = ((df['Bowler'] == player) & (~df['Wicket'].isin(['run out', 'obstructing the field'])))
    
    return df[filt]['Wicket'].count()


top_wicket_takers = pd.DataFrame()

top_wicket_takers['Players'] = pd.Series(df['Bowler'].unique())
top_wicket_takers['Innings Played'] = top_wicket_takers['Players'].apply(lambda player: df[df['Bowler'] == player]['Match ID'].unique().size)
top_wicket_takers['Wickets'] = top_wicket_takers['Players'].apply(get_wicket_count)
top_wicket_takers = top_wicket_takers.sort_values(by=['Wickets', 'Players'], ascending=[False, True])[:10]
top_wicket_takers
```

```python
def get_runs_conceded(player):
    filt = ((df['Bowler'] == player) & (df['Leg Byes'] == 0) & (df['Byes'] == 0))
    
    return df[filt]['Total Runs'].sum()


def get_overs_bowled(player):
    filt = ((df['Bowler'] == player) & (df['Wides'] == 0) & (df['No Balls'] == 0))
    
    balls_count = df[filt]['Ball'].count()
    return ((balls_count // 6) + (0.1 * (balls_count % 6)))


top_wicket_takers['Runs'] = top_wicket_takers['Players'].apply(get_runs_conceded)
top_wicket_takers['Overs'] = top_wicket_takers['Players'].apply(get_overs_bowled)
top_wicket_takers['Average'] = top_wicket_takers['Runs'] / top_wicket_takers['Wickets']
balls_delivered = top_wicket_takers['Overs'].astype(str).str.split('.').apply(lambda val: int(val[0]) * 6 + int(val[1]))
overs_propotionate = top_wicket_takers['Overs'].astype(str).str.split('.').apply(lambda val: float(val[0]) + (float(val[1]) / 6))
top_wicket_takers['Economy Rate'] = top_wicket_takers['Runs'] / overs_propotionate
top_wicket_takers['Strike Rate'] =  balls_player / top_wicket_takers['Wickets']
top_wicket_takers
```

<b>Economy v/s Wickets</b>
<br>
While the number of fast bowlers and spin bowlers in the list were same, what we find is that the economy rate for spinners was compartively lower and so were the number of wickets. Now this could be because of fast bowlers bowling more in powerplay and death overs, the two phases of the game where runs are scored at a higher pace. It is not something we can be sure as of now, since I don't have the data to back our claim, but that's what the rough analysis of the game say.

<b>Home and Away Wickets</b>
<br>
The qualifiers were played at different places, so considering them for this comparison was not fair. Hence only play offs were considered. Picking up your bowlers to suit your home advantage is a big part of team building and CSK here gives an example of how to use that home advantage, their spin department leads the charge at home where they have spin friendly wickets with Harbhajan Singh having the biggest difference between home and away wickets. On the opposite end we Kagiso Rababda who did exceptionally well on away matches with 64% of his wickets coming from away matches. 7 out of 10 bowlers performed better in their home conditions, however another factor to look into is, if qualifiers are brought into the picture the difference between Home and Away seems to even out for Imran Tahir and Deepak Chahar

```python
def get_wicket_count_type(player, venue_type):
    play_offs = match_df.iloc[:56]
    team = df[df['Bowler'] == player].iloc[0, 5]
    home_m_ids = play_offs[play_offs[venue_type] == team]['Match ID']
    home_matches = df[df['Match ID'].isin(home_m_ids)]
    
    filt = ((home_matches['Bowler'] == player) & (~home_matches['Wicket'].isin(['run out', 'obstructing the field'])))
    
    return home_matches[filt]['Wicket'].count()

top_wicket_takers['Home Wickets'] = top_wicket_takers['Players'].apply(get_wicket_count_type, args=('Team One',))
top_wicket_takers['Away Wickets'] = top_wicket_takers['Players'].apply(get_wicket_count_type, args=('Team Two',))
top_wicket_takers
```
